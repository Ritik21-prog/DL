```python
def check(upperb, lowerb, ip):
  lower_bound = lowerb
  upper_bound = upperb

  def label_number(number):
    if number >= lower_bound and number <= upper_bound:
        return 1
    else:
        return 0

  input_numbers = np.random.randint(0, 300, 1000)
  labels = [label_number(number) for number in input_numbers]

    # Split the dataset into training and testing sets
  split_ratio = 0.8
  split_index = int(split_ratio * len(input_numbers))
  train_X, train_y = input_numbers[:split_index], labels[:split_index]
  test_X, test_y = input_numbers[split_index:], labels[split_index:]
  # Train a decision tree classifier on the training set
  classifier = DecisionTreeClassifier()
  classifier.fit(train_X.reshape(-1, 1), train_y)

  # Evaluate the performance of the classifier on the testing set
  accuracy = classifier.score(test_X.reshape(-1, 1), test_y)
  print("Accuracy:", accuracy)

    # Predict whether a new input number lies between the two given numbers
  new_number = ip
  prediction = classifier.predict([[new_number]])
  prediction = int(prediction)
  return prediction
```

```python
input = [34, 60, 80, 90, 100, 120]
```

```python
# Apply the function to all elements of the list using a lambda function
predict = list(map(lambda x,y,z: check(x,y,z), max, min, input))
```
